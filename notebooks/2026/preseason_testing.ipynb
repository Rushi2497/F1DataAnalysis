```python
test_data = {}
all_laps = []

for i in range(1,3):
    for j in range(1,4):
        test_data[(i, j)] = fastf1.get_testing_session(year=2026, test_number=i, session_number=j)
        test_data[(i, j)].load()
        all_laps.append(test_data[(i, j)].laps)
```
